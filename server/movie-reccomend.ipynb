```python
movies.rename(columns={'id':'movieId'}, inplace=True)
```

```python
movies = movies[['movieId','imdb_id','title','vote_average','vote_count','poster_path']]
movies.head()
```

```python
check = ratings.merge(links, on='movieId')
check.head()
```

```python
movies = movies[movies['movieId'] != '2014-01-01']
movies = movies[movies['movieId'] != '1997-08-20']
movies = movies[movies['movieId'] != '2012-09-29']
```

```python
movies['movieId']=movies['movieId'].astype(int)
```

```python
rating_by_name = movies.merge(ratings, on='movieId')
rating_by_name.drop(['vote_average','vote_count'], axis=1, inplace=True)
rating_by_name
```

```python
x = rating_by_name.groupby('userId').count()['rating'] > 100
good_users = x[x].index
```

```python
mid_df = rating_by_name[rating_by_name['userId'].isin(good_users)]
mid_df.shape
```

```python
y = mid_df.groupby('title').count()['rating']>25
good_movies = y[y].index
```

```python
final_df = mid_df[mid_df['title'].isin(good_movies)]
```

```python
final_movies = final_df.drop(['timestamp'], axis=1)
```

```python
final_movies.head()
```

```python
movies_pivot = final_movies.pivot_table(index='title', columns='userId', values='rating')
movies_pivot
```

```python
movies_pivot.fillna(0,inplace=True)
```

```python
movies_pivot.head(150)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_mat = cosine_similarity(movies_pivot)
```

```python
similarity_mat.shape
```

```python
#array
similarity_mat
```

##FUNCTION TO RECOMMEND 6 MOVIES

```python
import numpy as np
def recommend(movie_name):
    index = np.where(movies_pivot.index == movie_name)[0][0]
    recommends = sorted(list(enumerate(similarity_mat[index])), key=lambda x:x[1], reverse=True)[1:6]
    data=[]
    for i in recommends:
        l=[]
        m = movies[movies['title'] == movies_pivot.index[i[0]]]
        # print(m.drop_duplicates('title')['title'].values)
        l.extend(list(m.drop_duplicates('title')['title'].values))
        l.extend(list(m.drop_duplicates('title')['movieId'].values))
        l.extend(list(m.drop_duplicates('title')['imdb_id'].values))
        l.extend(list(m.drop_duplicates('title')['poster_path'].values))
        data.append(l)
    print(data)
```

```python
recommend('Top Gun')
```
